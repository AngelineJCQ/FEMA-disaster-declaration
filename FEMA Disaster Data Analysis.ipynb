# Part 3: Area Trend

```python
df_area = df.groupby(by=['state'])['disasterNumber'].count().sort_values(ascending=False)
df_area.head(10)
```

```python
df_nocovid
```

```python
statesData = pd.pivot_table(df_nocovid, 
                            index='state', 
                            values='femaDeclarationString', 
                            columns='incidentType', aggfunc='count').reset_index()
```

```python
statesData = statesData.rename_axis(None, axis=1)
statesData.fillna(0, inplace=True)
```

```python
disaster = statesData.columns.values.tolist()
```

```python
statesData['sum'] = statesData[disaster[1:]].sum(axis=1)
statesData['dominant'] = statesData[disaster[1:]].idxmax(axis=1)
statesData['dominant'] = statesData['dominant'].str.title()
```

```python
statesData.head()
```

```python
countiesData = df_nocovid.copy(deep=True)
countiesData['fipsCountyCode'] = countiesData['fipsCountyCode'].astype('int').astype('str').str.zfill(3)
countiesData['fipsStateCode'] = countiesData['fipsStateCode'].astype('str')
countiesData['GEOID'] = countiesData['fipsStateCode'] + countiesData['fipsCountyCode']
```

```python
countiesData
```

```python
countiesData = pd.pivot_table(countiesData, 
                            index='GEOID', 
                            values='femaDeclarationString', 
                            columns='incidentType', aggfunc='count').reset_index()
countiesData = countiesData.rename_axis(None, axis=1)
countiesData.fillna(0, inplace=True)
```

```python
countiesData['sum'] = countiesData[disaster[1:]].sum(axis=1)
countiesData['dominant'] = countiesData[disaster[1:]].idxmax(axis=1)
countiesData['dominant'] = countiesData['dominant'].str.title()
```

```python
countiesData.head()
```

```python
import geopandas as gpd
counties = gpd.read_file('https://www2.census.gov/geo/tiger/GENZ2020/shp/cb_2020_us_county_5m.zip')
states = gpd.read_file('https://www2.census.gov/geo/tiger/GENZ2020/shp/cb_2020_us_state_5m.zip')
```

```python
statesData.head()
```

```python
statesData['state'] = statesData['state'].str.title()
states['STUSPS'] = states['STUSPS'].str.title()
statesData.rename(columns={'state':'STUSPS'}, inplace=True)
statesDisasters = states.merge(statesData, on='STUSPS')
```

```python
countiesDisasters = counties.merge(countiesData, on='GEOID')
```

```python
countiesDisasters.to_file('countiesDisasters.geojson', driver='GeoJSON')
statesDisasters.to_file('statesDisasters.geojson', driver='GeoJSON')
```
